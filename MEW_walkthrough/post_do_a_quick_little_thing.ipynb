```python
import networkx as nx
import json 
```

```python
filename = 'SC/SC_Processed_Precincts.json'

with open(filename) as f:
    data = json.load(f)

g = nx.json_graph.adjacency_graph(data)

g = nx.convert_node_labels_to_integers(g, ordering='default', label_attribute='GEOID')

data = nx.node_link_data(g)
with open('SC/SC_dual_graph_stripped.json', 'w') as f:
    json.dump(data, f)
```
