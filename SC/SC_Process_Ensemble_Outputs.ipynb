```python
df = pd.read_csv("./SC_Ensemble_Stats/chain_outputs_0_1000.csv")

for t in ts[1:]:
    df2 = pd.read_csv(f"./SC_Ensemble_Stats/chain_outputs_0_{t}.csv")
    df = pd.concat([df, df2], ignore_index=True)
```

```python
counts = Counter(df['County splits'])

categories = list(counts.keys())
frequencies = list(counts.values())
plt.bar(categories, frequencies)
plt.axvline(10, color='red',label='Enacted')
plt.title("County splits")
plt.show()

plt.hist(df['MM'],bins=100) 
plt.axvline(-0.02801788672558564, color='red',label='Enacted')
plt.title("Mean median")
plt.show()

plt.hist(df['EG'],bins=100)
plt.axvline(-0.19522613075328088, color='red',label='Enacted')
plt.title("Efficiency gap")
plt.show()

plt.hist(df['PB'])
plt.axvline(-0.07142857142857145, color='red',label='Enacted')
plt.title("Partisan bias")
plt.show()

plt.hist(df['PP'],bins=100)
plt.axvline(2.9564023402099537, color='red',label='Enacted')
plt.title("Polsby popper")
plt.show()

counts2 = Counter(df['DWins'])

categories = list(counts2.keys())
frequencies = list(counts2.values())

plt.bar(categories, frequencies)
plt.axvline(1, color='red',label='Enacted')
plt.title("Dem wins")
plt.show()
```
